<h3>&sect;2.2: Probleemdefinitie en Doelstelling</h3>

Connect Four is een strategisch spel voor twee spelers waarin de spelers om de beurt een schijfje (beide een eigen kleur) in een van de zeven kolommen laten vallen. Het doel is om als eerste vier de zelfde kleur schijven op een rij te krijgen (horizontaal, verticaal of diagonaal).

Voor deze opdracht willen we een rule-based agent ontwikkelen die effectief tegen een menselijke tegenstander kan spelen. De uitdaging is om een set regels te definiëren die de agent in staat stelt om strategische zetten te maken en te reageren op acties van de tegenstander.

De belangrijkste problemen die we willen aanpakken zijn:
- ***Bepalen van de beste zet***: Hoe kan de agent op basis van regels een sterke zet kiezen?
- ***Voorkomen van verlies***: Hoe kan de agent verlies vermijden door tegen de strategie van de tegenstander in te spelen?
- ***Balans tussen aanvallen en verdedigen***: Hoe kan de agent slim kiezen tussen het opbouwen van een eigen winnende combinatie en het blokkeren van de tegenstander?

Een rule-based systeem is hier een geschikte aanpak om de volgende redenen:
- ***Transparantie en begrijpelijkheid***: In tegenstelling tot machine learning-modellen zijn rule-based systemen goed te begrijpen en uitlegbaar. We kunnen precies zien welke regels de agent volgt en waarom een bepaalde zet wordt gemaakt.
- ***Snelheid en efficiëntie***: Een rule-based agent kan snelle beslissingen nemen zonder dat er complexe berekeningen of trainingstijd nodig is. Dit is handig voor real-time spelinteracties.
- ***Duidelijke strategieën en patronen***: Veel succesvolle Connect Four-strategieën kunnen worden vertaald naar regels, zoals het blokkeren van een tegenstander met drie op een rij of het benutten van een vrije kolom voor een winnende zet.
- ***Minder rekenkracht nodig***: In vergelijking met machine learning-modellen zoals deep reinforcement learning, vereist een rule-based systeem enn stuk minder rekenkracht en geheugen.

Een alternatieve aanpak zoals machine learning (bijv. een reinforcement learning agent) zou mogelijk sterker kunnen zijn, maar heeft een lange trainingsperiode nodig en vereist een grote hoeveelheid data. Voor deze opdracht is een rule-based aanpak dus een logische keuze.

```python
env = connect_four_v3.env(render_mode="rgb_array")
env.reset(seed=42)

frames = []

for agent in env.agent_iter():
    observation, reward, termination, truncation, info = env.last()

    if termination or truncation:
        action = None
    else:
        mask = observation["action_mask"]
        action = env.action_space(agent).sample(mask)

    env.step(action)

    frame = env.render()
    frames.append(frame)

    plt.imshow(frame)
    plt.axis("off")
    clear_output(wait=True)
    display(plt.gcf())
    time.sleep(0.5)

env.close()
```

1. Basisregels

Deze regels bepalen het meest fundamentele gedrag van de agent.
- Als de agent een zet kan doen waarmee hij direct wint, doe dan die zet.
- Als de tegenstander een zet kan doen waarmee hij wint, blokkeer die zet.
- Als er geen directe win- of verlieszet is, speel een zet die toekomstige winstmogelijkheden creëert.
- Als de agent niet strategisch hoeft te spelen, geef dan voorkeur aan het spelen in de middelste kolommen (indien beschikbaar).

Naast de basisregels zijn er ook nog prioriteiten en strategische regels nodig om het spel uiteindelijk te kunnen winnen. Er zijn hier ook verschillende lagen in, omdat er vaak meerdere mogelijke opties zijn, maar sommige zetten een hogere prioriteit hebben dan de andere. Zo is de zet met de hoogste prioriteit een winnende zet plaatsen. Deze heeft voorrang op een regel als "Het blokkeren van de vier op een rij van de tegenstander". Met de winnende zet is het spel namelijk afgelopen en heb je ook nog eens gewonnen. Hieronder worden de regels verdeeld onder vier soorten prioriteit van hoog naar middelhoog naar laag. Als er meerdere regels van toepassing zijn, moet de agent ze in een logische volgorde verwerken. Hieronder staan de regels in volgorde van belangrijkheid:

2. Prioriteiten en Strategische Regels

**Win- en Blokkeerregels (hoogste prioriteit)**
- *Regel 1*: Als de agent een zet kan doen waarmee hij direct vier op een rij krijgt, doe deze zet.
- *Regel 2*: Als de tegenstander een zet kan doen waarmee hij direct vier op een rij krijgt, blokkeer die zet.

**Aanvalsstrategieën (middelhoge prioriteit)**
- *Regel 3*: Probeer drie op een rij te maken met een open einde (zodat je later kunt winnen).
- *Regel 4*: Probeer een dubbele dreiging te creëren (twee mogelijke win-situaties tegelijk).
- *Regel 5*: Speel in een kolom waar je later een winnende zet kunt forceren.

**Verdedigingsstrategieën (middelhoge prioriteit)**
- *Regel 6*: Blokkeer tegenstander als hij drie op een rij heeft met een open einde.
- *Regel 7*: Vermijd zetten die de tegenstander een voordeel geven (bijv. een valstrik zetten waarin hij gegarandeerd wint).

**Positionele Voorkeuren (lage prioriteit)**
- *Regel 8*: Geef de voorkeur aan het spelen in de middelste kolommen boven de buitenste.
- *Regel 9*: Vermijd zetten in kolommen die de tegenstander helpen bij zijn strategie.

**Rule based action logged**

Deze functie is eigenlijk een wrapper voor de bestaande rule_based_action() uit main.py. **Waarom doen we dit?**

Zo kun je later precies zien welke regel er gebruikt werd, wat handig is om te checken of de AI op de juiste manier beslissingen maakt.

Als je in de toekomst meer info wil toevoegen aan je log, kun je dat makkelijk in deze wrapper doen zonder de originele code in main.py aan te passen.

***Consistentie:***
Alle AI-zetten worden nu altijd teruggegeven als een tuple, wat het loggen en analyseren een stuk eenvoudiger maakt.

**Check Win**

Deze functie is toegevoegd omdat ons systeem (via main.py) moeite had om de winnaar correct te bepalen tijdens het simuleren van de spellen. Soms werd er simpelweg geen winnaar aangegeven, en werd de agent gewoon "terminated". Daarom maken we met check_win_local() zelf een inspectie van het bord om te kijken of er vier opeenvolgende steentjes van één speler staan.

---

***Wat doet de functie precies?***

---

Door deze eigen check kunnen we met zekerheid bepalen of speler 1 of speler 2 vier-op-een-rij heeft en zo de winnaar vaststellen, in plaats van te vertrouwen op de reward-signalen van de omgeving.

***Horizontaal:*** Het doorloopt elke rij en checkt of er vier opeenvolgende steentjes van de gegeven speler staan.

***Verticaal:*** Het doorloopt elke kolom en controleert hetzelfde.

***Diagonaal:*** Er zijn twee diagonaal-richtingen: van linksboven naar rechtsonder en van rechtsboven naar linksonder. In beide richtingen wordt gekeken of er vier opeenvolgende steentjes van de speler staan.


```python
def check_win_local(board, player):
    """
    Controleert of de gegeven speler (1 of 2) vier-op-een-rij heeft op het board.
    board: 2D-array (6x7), met 1 = player_0, 2 = player_1
    player: 1 of 2
    """
    # Horizontaal
    for r in range(6):
        for c in range(7 - 3):
            if all(board[r, c + i] == player for i in range(4)):
                return True
    # Verticaal
    for r in range(6 - 3):
        for c in range(7):
            if all(board[r + i, c] == player for i in range(4)):
                return True
    # Diagonaal (links-boven naar rechts-onder)
    for r in range(6 - 3):
        for c in range(7 - 3):
            if all(board[r + i, c + i] == player for i in range(4)):
                return True
    # Diagonaal (rechts-boven naar links-onder)
    for r in range(6 - 3):
        for c in range(3, 7):
            if all(board[r + i, c - i] == player for i in range(4)):
                return True
    return False
```

***convert observation swapped***

Deze functie is toegevoegd omdat we met het standaard main.convert_observation() niet goed konden bepalen wie er gewonnen had. De observaties die we krijgen van de PettingZoo-omgeving geven de posities van de stenen via twee kanalen (één per speler), maar de volgorde van die kanalen komt niet overeen met hoe wij de spelers willen labelen.

Met convert_observation_swapped() doen we het volgende:

We zeggen dat als er een '1' staat in kanaal 1 van de observatie, dat hoort bij player_0 (en geven dit op het bord als waarde 1).
Als er een '1' staat in kanaal 0, dan hoort dat bij player_1 (en zetten we dat op het bord als 2).
Door deze omwisseling (swapping) zorgen we ervoor dat het interne bord (de 2D-array) correct weergeeft welke zet door welke speler is gemaakt. Hierdoor kan onze check_win_local() functie de juiste winnaar bepalen, omdat de interne waarden nu overeenkomen met onze definitie van speler 1 en speler 2.

```python
def convert_observation_swapped(obs):
    """
    Aangepaste versie van main.convert_observation().
    We gaan er in dit voorbeeld vanuit dat:
    - player_0 zit in obs[..., 1] => dat wordt '1' op het bord
    - player_1 zit in obs[..., 0] => dat wordt '2' op het bord

    Pas dit aan als jouw tests laten zien dat de mapping anders moet zijn!
    """
    import numpy as np

    board = np.zeros((6, 7), dtype=int)
    for i in range(6):
        for j in range(7):
            # Als kanaal 1 == 1 => player_0 => op het bord = 1
            if obs[i, j, 1] == 1:
                board[i, j] = 1
            # Als kanaal 0 == 1 => player_1 => op het bord = 2
            elif obs[i, j, 0] == 1:
                board[i, j] = 2
    return np.flipud(board)
```

<div style="background-color:#506375; color:white; padding: 10px;"> <strong>Uitvoeren van een singulair gegenereerd spel</strong> </div>
Deze stap was voor mij vooral voor het bugfixen van de code, hier kan je zien of het rule based systeem wel de correcte keuzes aan het maken was, tijdens het testen van deze functie kreeg ik namelijk vaak dat de keuzes van het rule based systeem erg verschilde met de keuzes die hij maakte als ik tegen het systeem speelde

```python
def display_single_game():
    """
    Speelt één spel en geeft na elke zet het spelbord weer.
    Logt de zetten en controleert na elke zet zelf of er een winnaar is.
    Geeft een nette boodschap: "Speler X heeft gewonnen, spel is voorbij" of "Gelijkspel, spel is voorbij".
    """
    # We resetten de PettingZoo-omgeving met een vaste seed
    main.env.reset(seed=42)
    
    print("Start van het spel:\n")
    main.render_board()

    # Mapping van PettingZoo-agentnamen naar interne speler-nummers (1 en 2).
    # We noemen player_0 => 1 en player_1 => 2.
    agent_to_speler = {"player_0": 1, "player_1": 2}
    
    # Variabelen om de zetten te loggen en de winnaar bij te houden
    game_moves = []
    winner = None

    # Doorloop de agenten in beurtvolgorde
    for agent in main.env.agent_iter():
        raw_observation, reward, termination, truncation, info = main.env.last()

        # Als deze agent al "dood" is (termination/truncation),
        # doen we een None-actie en stoppen we dit spel
        if termination or truncation:
            main.env.step(None)
            print(f"\n[{agent}] Spel beëindigd!")
            break

        # === Aangepaste convert_observation() om de kanalen te swappen ===
        observation = {
            "board": convert_observation_swapped(raw_observation["observation"]),
            "action_mask": raw_observation["action_mask"]
        }
        # =================================================================

        # Bepaal welke zet wordt gedaan
        if agent == "player_0":
            move, rule = rule_based_action_logged(observation, agent)
        else:
            # Speler 2: check eerst of kolom 3 vrij is, anders willekeurig
            mask = observation["action_mask"]
            if mask[3]:
                move = 3
                rule = "Start in het midden (Speler 2)"
            else:
                valid_moves = [i for i, valid in enumerate(mask) if valid]
                move = random.choice(valid_moves)
                rule = "Willekeurige zet (Speler 2)"
        
        # Log deze zet
        game_moves.append({"agent": agent, "move": move, "rule": rule})
        print(f"Speler {agent_to_speler[agent]} zet: {move} | Toegepaste regel: {rule}")
        
        # Voer de zet uit in de PettingZoo-omgeving en render het bord
        main.env.step(move)
        main.render_board()
        
        # Haal de bijgewerkte observatie op, converteer met de swapped-functie
        updated_raw_obs, _, _, _, _ = main.env.last()
        updated_board = convert_observation_swapped(updated_raw_obs["observation"])

        # Bepaal de interne speler (1 of 2)
        current_player = agent_to_speler[agent]

        # Check of deze speler gewonnen heeft
        if check_win_local(updated_board, current_player):
            winner = current_player
            print(f"\nSpeler {winner} heeft gewonnen, spel is voorbij.")
            break

        # Als er geen zetten meer mogelijk zijn => gelijkspel
        if not any(updated_raw_obs["action_mask"]):
            winner = "Gelijkspel"
            print("\nGelijkspel, spel is voorbij.")
            break

    # Als we geen winnaar gevonden hebben, is het een gelijkspel
    if winner is None:
        winner = "Gelijkspel"

    print("\nEindtoestand:")
    main.render_board()
    
    if winner == "Gelijkspel":
        print("\nGelijkspel, spel is voorbij.")
    else:
        print(f"\nSpeler {winner} heeft gewonnen, spel is voorbij.")
        
    return game_moves

if __name__ == "__main__":
    moves = display_single_game()
    print("\nGemaakte zetten in dit spel:")
    for m in moves:
        print(m)
```

```python
def simulate_games_with_images(num_games=25):
    """
    Speelt num_games spellen en slaat voor elk spel het laatste frame (RGB-afbeelding) op.
    Spelen tussen:
      - player_0: rule-based strategie (via rule_based_action_logged)
      - player_1: vaste strategie (start in het midden indien mogelijk, anders willekeurig)
    Retourneert:
      - wins: dict met aantal winsten voor {1, 2, "draw"}
      - final_frames: lijst met RGB-afbeeldingen (het laatste frame van elk spel)
    """
    wins = {1: 0, 2: 0, "draw": 0}
    final_frames = []
    agent_to_speler = {"player_0": 1, "player_1": 2}

    for game_idx in range(num_games):
        main.env.reset(seed=random.randint(0, 999999))
        winner = None

        # Doorloop de agenten met een for-loop (dit doorloopt alle beurten in het spel)
        for agent in main.env.agent_iter():
            raw_observation, reward, termination, truncation, info = main.env.last()

            # Als het spel is beëindigd, voer dan step(None) uit en stop deze game
            if termination or truncation:
                main.env.step(None)
                break

            observation = {
                "board": convert_observation_swapped(raw_observation["observation"]),
                "action_mask": raw_observation["action_mask"]
            }

            # Kies de zet
            if agent == "player_0":
                move, rule = rule_based_action_logged(observation, agent)
            else:
                mask = observation["action_mask"]
                if mask[3]:
                    move = 3
                    rule = "Start in het midden (Speler 2)"
                else:
                    valid_moves = [i for i, valid in enumerate(mask) if valid]
                    move = random.choice(valid_moves)
                    rule = "Willekeurige zet (Speler 2)"

            main.env.step(move)

            # Haal de bijgewerkte observatie op en check of de huidige speler gewonnen heeft
            updated_raw_obs, _, _, _, _ = main.env.last()
            updated_board = convert_observation_swapped(updated_raw_obs["observation"])
            current_player = agent_to_speler[agent]
            if check_win_local(updated_board, current_player):
                winner = current_player
                break

            # Als er geen geldige zetten meer zijn, beschouw als gelijkspel
            if not any(updated_raw_obs["action_mask"]):
                winner = "draw"
                break

        if winner is None:
            winner = "draw"

        # Update wins-teller
        if winner == 1:
            wins[1] += 1
        elif winner == 2:
            wins[2] += 1
        else:
            wins["draw"] += 1

        # Sla het laatste frame van dit spel op als visualisatie
        final_frame = main.env.render()
        final_frames.append(final_frame)

    return wins, final_frames

def plot_final_frames_grid(final_frames, grid_shape=(5, 5)):
    """
    Maakt een grid-visualisatie (bijv. 5x5) van de laatste frames van de spellen.
    final_frames: lijst met RGB-afbeeldingen.
    grid_shape: tuple (n_rows, n_cols)
    """
    n_rows, n_cols = grid_shape
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 3, n_rows * 3))
    fig.suptitle("Laatste frames van de spellen", fontsize=16)
    for idx, ax in enumerate(axes.flatten()):
        if idx < len(final_frames):
            ax.imshow(final_frames[idx])
            ax.axis("off")
        else:
            ax.axis("off")
    plt.tight_layout()
    plt.show()
```

<div style="background-color:#506375; color:white; padding: 10px;"> <strong>Uitkomsten van 25 spellen laten zien</strong> </div>

```python
wins, final_frames = simulate_games_with_images(25)
print("Resultaten over 25 spellen:")
print(f"Speler 1 (player_0) wint: {wins[1]} keer")
print(f"Speler 2 (player_1) wint: {wins[2]} keer")
print(f"Gelijkspel: {wins['draw']} keer")
    
plot_final_frames_grid(final_frames, grid_shape=(5, 5))
```
